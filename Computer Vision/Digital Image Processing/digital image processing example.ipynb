Here, Iam providing the basic examples of digital image processing for beginners, including loading the image, converting it into grayscale, performing edge detection, and more.

#### Import Necessary libraries

Here we need to import 'cv2', which is used for image processing tasks, and 'matplotlib', which is used for visualization.

```python
import cv2
import matplotlib.pyplot as plt
```

#### üñºÔ∏è Load the image path and read the image


```python
# Load the image
image_path = "C:/Users/gokul/OneDrive/Pictures/IMG_4221_edited.jpg"  #specify the image path
image = cv2.imread(image_path)                                      
```

#### üåà Convert the image into grayscale with cv2.COLOR_BGR2GRAY ‚ö™‚ö´

```python
# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

```

#### Blur the image üå´Ô∏è

```python
# Apply Gaussian blur to the grayscale image
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)
```

#### Apply Edge detection with Canny üîç

```python
# Perform edge detection using Canny
edges = cv2.Canny(blurred_image, 50, 150)
```

#### Perform Thresholding on the blurred image

```python
# Thresholding
_, thresholded_image = cv2.threshold(blurred_image, 100, 255, cv2.THRESH_BINARY)
```

#### Apply Contour detection 

 It is a technique used in image processing to find and outline the boundaries of objects in an image

```python
# Contour detection
contours, _ = cv2.findContours(edges.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

```

```python
# Draw contours on the original image
contour_image = image.copy()
cv2.drawContours(contour_image, contours, -1, (0, 255, 0), 2)
```

#### Visualise all the actions performed

```python
# Display the images
plt.figure(figsize=(16, 8))

# Original Image
plt.subplot(2, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Grayscale Image
plt.subplot(2, 3, 2)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

# Blurred Image
plt.subplot(2, 3, 3)
plt.imshow(blurred_image, cmap='gray')
plt.title('Blurred')
plt.axis('off')


# Edge Detected Image
plt.subplot(2, 3, 4)
plt.imshow(edges, cmap='gray')
plt.title('Edge Detected Image')
plt.axis('off')

# Thresholded Image
plt.subplot(2, 3, 5)
plt.imshow(thresholded_image, cmap='gray')
plt.title('Thresholded')
plt.axis('off')

# Contour Detected Image
plt.subplot(2, 3, 6)
plt.imshow(cv2.cvtColor(contour_image, cv2.COLOR_BGR2RGB))
plt.title('Contours')
plt.axis('off')

plt.tight_layout()
plt.show()

```
